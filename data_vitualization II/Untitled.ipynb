```python
# Data vitualization II
# Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution 
# of age with respect to each gender along with the information about whether they survived 
# or not. (Column names : 'sex' and 'age') 
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the Titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic
```

```python
# Plot a box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic)
plt.title('Distribution of Age with Respect to Gender and Survival')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
# Plot a box plot without outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x='sex', y='age', hue='survived', data=titanic, showfliers=False)
plt.title('Distribution of Age with Respect to Gender and Survival (Without Outliers)')
plt.xlabel('Gender')
plt.ylabel('Age')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```
