```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the Titanic dataset
titanic_df = sns.load_dataset('titanic')
```

```python
# Display the first few rows of the dataset
print("First few rows of the Titanic dataset:")
print(titanic_df.head())
```

```python
# Drop categorical variables
numerical_df = titanic_df.drop(['sex', 'embarked', 'class', 'who', 'deck', 'embark_town', 'alive', 'adult_male', 'alone'], axis=1)
```

```python
# Plot a countplot of survival
plt.figure(figsize=(6, 4))
sns.countplot(x='survived', data=titanic_df)
plt.title('Count of Passengers by Survival Status')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.show()
```

```python
# Use Seaborn to create a pairplot
sns.pairplot(titanic_df, hue='survived', palette='Set1')
plt.title('Pairplot of Titanic Dataset')
plt.show()
```

```python
# Use Seaborn to create a heatmap to visualize correlations
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap of Titanic Dataset (Numerical Variables)')
plt.show()
```

```python
# Write a code to check how the price of the ticket (column name: 'fare') for each passenger 
# is distributed by plotting a histogram.
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=titanic_df, x='fare', bins=20, kde=True, color='skyblue')
plt.title('Distribution of Ticket Prices (Fare)')
plt.xlabel('fare')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
