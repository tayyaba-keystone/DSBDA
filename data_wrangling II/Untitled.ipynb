```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('Academic.csv')
```

```python
df
```

```python
df.isnull().sum()
```

```python
df['Email']=df['Email'].fillna('Keystone_student')
```

```python
df
```

```python
# plt.figure(figsize=(10, 6))
df.boxplot()
plt.title('Boxplot of DataFrame Features (With Outliers)')
plt.ylabel('Values')
plt.show()
```

```python
# Create a boxplot without outliers
# plt.figure(figsize=(10, 6))
df.boxplot(showfliers=False)
plt.title('Boxplot of DataFrame Features (Without Outliers)')
plt.ylabel('Values')
plt.show()
```

```python
# Plot histograms before and after transformation
plt.figure(figsize=(12, 6))
```

```python
# Histogram of 'First year: Sem 1' before transformation
plt.subplot(1, 2, 1)
plt.hist(df['PRN No.'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of "PRN NO." (Before Transformation)')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
# Histogram of 'PRN NO.' after transformation
plt.subplot(1, 2, 2)
plt.hist(df['PRN No.'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Histogram of "PRN NO." (After Transformation)')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```
